```python
x=mnist.iloc[:,1:]
y=mnist.iloc[:,0].to_frame()
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(mnist.iloc[:,1:],mnist.iloc[:,0].to_frame(),random_state=42,test_size=.2,stratify=y)
```

```python
# using knn_classifier
from sklearn.neighbors import KNeighborsClassifier

knn_clf=KNeighborsClassifier()

```

```python
from sklearn.model_selection import cross_val_score

cross_val_score(knn_clf,x_train,y_train,cv=3,scoring='accuracy')
```

Scaling the features

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.fit(x_test)
```

```python
#as range is less scaling not a good idea it could introduce noise
```

```python
knn_clf2=KNeighborsClassifier()
cross_val_score(knn_clf2,x_train_scaled,y_train,cv=3,scoring='accuracy')
```

```python
from sklearn.model_selection import GridSearchCV

grid_params=[{'n_neighbors':[3,4,5,6],'weights':["uniform","distance"]}]

knn_clf3=KNeighborsClassifier()
grid_search=GridSearchCV(knn_clf3,cv=5,param_grid=grid_params)
```

```python
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```

```python
acccuracy_from_best_estimator=grid_search.score(x_test,y_test)
acccuracy_from_best_estimator
```

Prediction 1

```python
test_set=pd.read_csv('digit-recognizer/test.csv')
test_set
```

```python
prediction=grid_search.predict(test_set)
prediction
```

```python
submission1=pd.DataFrame(
    {
        'ImageId':test_set.index+1,
        'Label':prediction
    }
)
submission1.set_index('ImageId',inplace=True)
submission1.to_csv('submission folder/submission1.csv')
```
